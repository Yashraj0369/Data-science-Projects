## Random forest

```python
X= bank_loan1.drop(columns=['Personal.Loan'])
y= bank_loan1['Personal.Loan']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=369)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(criterion='gini',max_depth=6,n_estimators=100,random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
predict = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(predict,y_test)
print(cm)
sns.heatmap(cm,annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predict,y_test)
print(accuracy)
```
